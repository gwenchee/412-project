```python
# get sqlite cursor and simulation time parameters
filename = '412project.sqlite'
conn = lite.connect(filename)
cur = conn.cursor()
cur = an.get_cursor(filename)
init_year, init_month, duration, timestep = an.get_timesteps(cur)
```

```python
# Uox waste per year (output from lwr uox)
uox_waste_amount = cur.execute('select quantity, Commodity, Time, transactions.resourceid FROM transactions INNER JOIN resources ON transactions.resourceid = resources.resourceid WHERE Commodity = "uox_waste"').fetchall()
list = []
for time in timestep:
    uox_waste =0
    for row in uox_waste_amount:
        if time ==row['time']:
            uox_waste = row['quantity']*1E-3
    list.append(uox_waste)
    
time = timestep/12 + init_year  

plt.bar(time, list)
plt.xlabel('Years')
plt.ylabel('Mass (MTHM)')
plt.title('uox waste produced per year')
plt.grid(color='grey', linestyle='-', linewidth=0.5)
plt.show()
```

```python
# Uox waste cumulative (output from lwr uox)
uox_waste_amount = cur.execute('select quantity, Commodity, Time, transactions.resourceid FROM transactions INNER JOIN resources ON transactions.resourceid = resources.resourceid WHERE Commodity = "uox_waste"').fetchall()
list = []
uox_waste = 0
for time in timestep:
    for row in uox_waste_amount:
        if time ==row['time']:
            uox_waste = uox_waste + row['quantity']*1E-3
    list.append(uox_waste)
        
time = timestep/12 + init_year    

plt.bar(time, list)
plt.xlabel('Years')
plt.ylabel('Mass (MTHM)')
plt.title('Cumulative Uox waste produced vs time')
plt.grid(color='grey', linestyle='-', linewidth=0.5)
plt.show()
```
